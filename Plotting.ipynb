## Simple plotting example

This is a plot of the $sinc$ function

$$
f(x) = \frac{\sin(x)}{x}
$$

```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)
plt.xticks([-4*pi, -2*pi, 0, 2*pi, 4*pi], ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'])
plt.plot(x, np.sin(x)/x)
plt.show()
```
